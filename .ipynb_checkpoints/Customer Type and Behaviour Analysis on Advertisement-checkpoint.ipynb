```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data= pd.read_csv('Clicked Ads Dataset.csv', index_col=0)
data.sample(2)
```

```python
data.shape
data.info()
```

```python
data.describe()
```

## Univariate

```python
num_dtypes = ['int64','float64']
num_df = data.select_dtypes(include=num_dtypes)
numerical = num_df.columns
print(numerical)
```

```python
plt.figure(figsize=(5,3))
for i in range (0, len(numerical)):
    plt.subplot(1, len(numerical), i+1)
    sns.boxplot(y=data[numerical[i]], color='green')
    plt.tight_layout()
```

```python
plt.figure(figsize=(5,3))
for i in range (0, len(numerical)):
    plt.subplot(2,2, i+1)
    sns.kdeplot(x=data[numerical[i]], color='green')
    plt.tight_layout()
```

```python
categori1 =['Male','Clicked on Ad']
plt.figure (figsize=(5,3))
for i in range (0, len(categori1)):
    plt.subplot(1,2 ,i+1)
    sns.countplot(x=data[categori1[i]],color='green')
    plt.tight_layout()
```

```python
categoria =['city','province','category']
plt.figure (figsize=(10,7))
for i in range (0, len(categoria)):
    plt.subplot(2,2 ,i+1)
    sns.countplot(x=data[categoria[i]],color='green')
    plt.tight_layout()
    plt.xticks(rotation=90)
```

```python
plt.figure(figsize = (18,5))
age = data['Age'].value_counts().reset_index()
sns.barplot(x='index', y='Age', data=age, orient='v')
plt.title('Distribution Age', weight='bold', fontsize=18)
plt.xlabel('')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(x='Daily Internet Usage', data=data);
```

```python
sns.barplot(y='Daily Internet Usage', hue='Clicked on Ad', data=data )
```

```python
sns.histplot(x='Daily Time Spent on Site',data=data)
```

```python
sns.histplot(x='province', data=data);
plt.xticks(rotation=90)
plt.xlabel('')
plt.ylabel('Value')

plt.show()
```
